# Loading User data

```python
query = '''select
                naid,
                game_net_id as platform_id,
                entrytag,
                install_date as timestamp,
                date(install_date) as day,
                hour(install_date) as hour,
                weekday(install_date) as weekday,
                country_id,
                channel_id
            from user_property
            where
                project_id = 107 and
                install_date between '2014-04-01' and '2014-04-29' '''

install_logs = kbn_sql.wrap_query('rkravitz', 'sqrt{nEg5}NOTufd', 'preprocessing_v3', query)
```

```python
install_logs.tail()
```

# Load in data from datameer: cityConstruction, rev, marches

```python
cityConstruction_7d = pd.read_csv(path + 'fte_agg_cityConstruction_7d.csv')
rev_7d = pd.read_csv(path + 'fte_agg_rev_7d.csv')
march_7d = pd.read_csv(path + 'fte_agg_march_7d.csv')
```

```python
cityConstruction_7d.install_date_min = pd.to_datetime(cityConstruction_7d.install_date_min)
cityConstruction_7d.install_date_min = pd.to_datetime(cityConstruction_7d.install_date_min.map(lambda x: x.date()))
cityConstruction_7d.head()

rev_7d.install_date = pd.to_datetime(rev_7d.install_date)
rev_7d.install_date = pd.to_datetime(rev_7d.install_date.map(lambda x: x.date()))
```

```python
march_7d = march_7d.pivot_table(rows = 'naid', cols = 'terrain', values = 'atk_cnt').reset_index()
```

```python
cityConstruction_7d.head(), rev_7d.head(), march_7d.head()
```

# Join Data to install logs

```python
df = install_logs.merge(cityConstruction_7d, on = 'naid', how = 'left').merge(rev_7d, on = 'naid', how = 'left').merge(march_7d, on = 'naid', how = 'left')
df = df[['naid', 'platform_id', 
         'entrytag', 'timestamp', 
         'day', 'hour', 
         'weekday', 'country_id', 
         'channel_id', 'cityConstruction_7d',
         'cityConstructionSum', 'maxPlayerLevel',
         'distDays',
         'rev', 'first_time', 
         'Forest', 'Mountain', 
         'Grassland']]
```

```python
df.day = pd.to_datetime(df.day)

entrytag_start = 'doah-pmv-501'
df['tag'] = df.entrytag.map(lambda x: x.startswith(entrytag_start))
```

```python
df.info()
```

```python
df.fillna(0, inplace = True)
```

# Quest Progression

```python
begin, end = pd.to_datetime('2014-04-18'), pd.to_datetime('2014-04-28')
df_during = df[(df.day >= begin) & (df.day <= end)]

df_during['Forested'] = df_during.Forest > 0
df_during['Mountained'] = df_during.Mountain > 0
df_during['Grasslanded'] = df_during.Grassland > 0

df_during.groupby('day')[['Forested', 'Mountained', 'Grasslanded']].mean().plot()
title('Quest Progression By Cohort Day')
show()
close()
(df_during.groupby('day').Mountained.mean() / df_during.groupby('day').Forested.mean()).plot()
(df_during.groupby('day').Grasslanded.mean() / df_during.groupby('day').Mountained.mean()).plot()
```

## In the US

```python
usa = df_during[df_during.country_id == 2]

usa.groupby('day')[['Forested', 'Mountained', 'Grasslanded']].mean().plot()
title('Quest Progression by Cohort Day, USA')
show()
close()
(usa.groupby('day').Mountained.mean() / usa.groupby('day').Forested.mean()).plot()
(usa.groupby('day').Grasslanded.mean() / usa.groupby('day').Forested.mean()).plot()
```

## Quest Progression Amongst Tagged Users

```python
#entrytag_start = 'doah-pmv-501'
#df_during['tag'] = df_during.entrytag.map(lambda x: x.startswith(entrytag_start))
tagged = df_during[df_during.tag]

tagged.groupby('day')[['Forested', 'Mountained', 'Grasslanded']].mean().plot()
title('Quest Progression by Cohort Day, Tagged Users')
show()
close()
(tagged.groupby('day').Mountained.mean() / tagged.groupby('day').Forested.mean()).plot()
(tagged.groupby('day').Grasslanded.mean() / tagged.groupby('day').Forested.mean()).plot()
```
